```python
import numpy as np
import matplotlib.pyplot as plt
print('starting...')
y = np.zeros([100, 10], dtype=float)
for j in range(0,10):
    file = 'results'+str(j+1)+'.txt'
    k = 0;
    with open('robustness remove/'+file, mode='r+') as f1:
        #print(file)
        for line in f1:
            #print(line)
            y[k][j] += float(line)
            k = k + 1
d_mean = np.zeros(100, dtype=float)
d_std = np.zeros(100, dtype=float)
for i in range(100):
    d_mean[i] = np.mean(y[i])
    d_std[i] = np.std(y[i])
w = 6
h = 5
d = 70
plt.figure(figsize=(w, h), dpi=d)
positions = np.arange(0, 1000,10)
#plt.bar(positions, d_mean, color="blue", yerr=d_std)
plt.errorbar(positions, d_mean, d_std, linestyle='solid', capsize=2)
plt.grid(True)
plt.xlim([0, 1000])
plt.xlabel("Number of iterations")
plt.ylim([0, 25])
plt.ylabel("Circularity Error")
#plt.title('Circularity error for 6 robots with target sensing \nand communication range of 5 meters')
#plt.show()
#plt.savefig('foo.png', bbox_inches='tight')

fig1 = plt.gcf()
plt.show()
plt.draw()
#plt.grid()
fig1.savefig('robustness.png', format='png', dpi=1200)
#fig1.savefig('tessstttyyy.png', dpi=d)

print('done...')
```
